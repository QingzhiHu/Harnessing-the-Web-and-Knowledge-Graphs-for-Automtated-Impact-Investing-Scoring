```python
def count_ranges(column_name):
    very_positive = 0
    positive = 0
    neutral = 0
    negative = 0
    very_negative = 0

    for element in msci[column_name].dropna().values:
        if element >= 5 and element <= 10:
            very_positive += 1

        if element >= 1 and element < 5:
            positive += 1

        if element > -1 and element < 1:
            neutral += 1

        if element <= -1 and element > -5:
            negative += 1

        if element <= -5 and element >= -10:
            very_negative += 1
            
    return [very_positive, positive, neutral, negative, very_negative]
        
    
```
