```python
def clean_content_all(x):
    character_list = []
    a_string = x
    a_string = a_string.rstrip("\n")
    a_string = a_string.replace("..", ".")
    a_string = a_string.replace("  ", ".")
    for character in a_string:
        if character.isalnum() or (character in [",", ".", " ", "-", "%", "!", ";", "(", ")"]):
            character_list.append(character)

    return "".join(character_list)

# clean pdf
import nltk
nltk.download('wordnet')
nltk.download('punkt')

import string
import re
from tqdm import tqdm
import pandas as pd
tqdm.pandas()
 
def extract_statements(text):
  
  # remove non ASCII characters
  printable = set(string.printable)
  text = ''.join(filter(lambda x: x in printable, text))
  
  lines = []
  prev = ""
  for line in text.split('\n'):
    # aggregate consecutive lines where text may be broken down
    # only if next line starts with a space or previous does not end with a dot.
    if(line.startswith(' ') or not prev.endswith('.')):
        prev = prev + ' ' + line
    else:
        # new paragraph
        lines.append(prev)
        prev = line
        
  # don't forget left-over paragraph
  lines.append(prev)
 
  # clean paragraphs from extra space, unwanted characters, urls, etc.
  # best effort clean up, consider a more versatile cleaner
  sentences = []
  
  for line in lines:
      # removing header number
      line = re.sub(r'^\s?\d+(.*)$', r'\1', line)
      # removing trailing spaces
      line = line.strip()
      # words may be split between lines, ensure we link them back together
      line = re.sub(r'\s?-\s?', '-', line)
      # remove space prior to punctuation
      line = re.sub(r'\s?([,:;\.])', r'\1', line)
      # ESG contains a lot of figures that are not relevant to grammatical structure
      line = re.sub(r'\d{5,}', r' ', line)
      # remove mentions of URLs
      line = re.sub(r'((http|https)\:\/\/)?[a-zA-Z0-9\.\/\?\:@\-_=#]+\.([a-zA-Z]){2,6}([a-zA-Z0-9\.\&\/\?\:@\-_=#])*', r' ', line)
      # remove multiple spaces
      line = re.sub(r'\s+', ' ', line)
      # remove multiple dot
      line = re.sub(r'\.+', '.', line)
      
      # split paragraphs into well defined sentences using nltk
      for part in nltk.sent_tokenize(line):
        sentences.append(str(part).strip())
 
  return sentences  

# from nltk import tokenize
# def extract_statements(text):
#     return tokenize.sent_tokenize(text)

# samsung2021 = dfs[dfs.company.str.contains("Samsung")].content.values[0]
# samsung2021 = clean_content_all(samsung2021)
# statements = extract_statements(samsung2021)
samsung = dfs[dfs.company.str.contains("Samsung")][["pdf_url","content","company"]]
samsung["content_clean"] = samsung["content"].progress_apply(clean_content_all)
samsung["statement"] = samsung["content_clean"].progress_apply(extract_statements)
samsung = samsung.explode('statement').reset_index(drop=True)

def split_sentence(text):
    return text.split(".")

samsung["statement"] = samsung["statement"].progress_apply(split_sentence)
samsung = samsung.explode('statement').reset_index(drop=True)
```

```python
%load_ext autoreload
%autoreload 2
# Used to import data from local.
import pandas as pd

# Used to create the dense document vectors.
import torch
from sentence_transformers import SentenceTransformer

# Used to create and store the Faiss index.
import faiss
import numpy as np
import pickle
from pathlib import Path

# Used to do vector searches and display the results.
from vector_engine.utils import vector_search, id2details
```

```python

# df = pd.DataFrame()
# df["statement"] = statements
# df["company"] = "Samsung Electronics Co Ltd"
df=samsung.dropna().copy()

def get_id(x):
    return int(x[1:])
    
# df["id"] = df["wikidata_id"].apply(get_id)
df["id"] = range(len(df))


# Instantiate the sentence-level DistilBERT
# model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')
# model = SentenceTransformer('all-distilroberta-v1 ')
model = SentenceTransformer('paraphrase-MiniLM-L3-v2')



# Check if GPU is available and use it
if torch.cuda.is_available():
    model = model.to(torch.device("cuda"))
print(model.device)
# Convert abstracts to vectors
embeddings = model.encode(df.statement.tolist(), show_progress_bar=True)


# Step 1: Change data type
embeddings = np.array([embedding for embedding in embeddings]).astype("float32")

# Step 2: Instantiate the index
index = faiss.IndexFlatL2(embeddings.shape[1])

# Step 3: Pass the index to IndexIDMap
index = faiss.IndexIDMap(index)

# Step 4: Add vectors and their IDs
index.add_with_ids(embeddings, 
                   df["id"].values.astype(np.int64))

print(f"Number of vectors in the Faiss index: {index.ntotal}")
df_keywords = pd.read_csv("SDG_key_words.csv")
tmp = df_keywords[["SDG 1"]].dropna().values.tolist()
user_query = " ".join([item for sublist in tmp for item in sublist])
D, I = vector_search([user_query], model, index, index.ntotal)
df_temp = df.set_index("id").loc[I[0]]
df_temp["rank"] = range(len(df_temp))
df_temp["relevance_score"] = D.flatten().tolist()
df_temp
```

```python
from urllib.parse import urlparse
def get_report_name(x):
    domain = urlparse(x).path.split("/")[-1]
    return domain
df_temp["report_name"] = df_temp["pdf_url"].progress_apply(get_report_name)
```

```python


question = '''How does Samsung helps poverty?"'''

paragraph = " ".join(df_rank[df_rank.label=="entail"].sort_values("prob", ascending = False)[["statement", "report_name", "label", "prob"]].statement.values[0:5])

encoding = tokenizer.encode_plus(text=question,text_pair=paragraph, add_special=True)

inputs = encoding['input_ids']  #Token embeddings
sentence_embedding = encoding['token_type_ids']  #Segment embeddings
tokens = tokenizer.convert_ids_to_tokens(inputs) #input tokens
```

```python
start_scores, end_scores = model(input_ids=torch.tensor([inputs]), token_type_ids=torch.tensor([sentence_embedding]))
start_index = torch.argmax(start_scores)

end_index = torch.argmax(end_scores)

answer = ' '.join(tokens[start_index:end_index+1])
answer
```
