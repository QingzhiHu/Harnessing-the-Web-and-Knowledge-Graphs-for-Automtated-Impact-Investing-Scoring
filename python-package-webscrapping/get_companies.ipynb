```python
fraction = 0.5
ids = list(set(data1[data1.count_numeric >= 60*fraction].csecid.values.tolist() + data2[data2.count_numeric >= 60*fraction].csecid.values.tolist()))
len(ids)
```

```python
companies = df[df.csecid.isin(ids)].name.unique().tolist()
# companies
```

```python
companies = [x for x in companies if "dead" not in x.lower()]
len(companies)
```
