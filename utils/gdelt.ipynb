```python
# Retrieve news
import urllib.request
from datetime import datetime
 
master_url = 'http://data.gdeltproject.org/gdeltv2/masterfilelist.txt'
master_file = urllib.request.urlopen(master_url)
min_date = datetime.strptime("2020 1 1 12", '%Y %m %d %H')
max_date = datetime.strptime("2020 1 5 12", '%Y %m %d %H')
to_download = []
for line in master_file:
    decoded_line = line.decode("utf-8")
    if 'gkg.csv.zip' in decoded_line:
        a = decoded_line.split(' ')
        file_url = a[2].strip()
        file_dte = datetime_object = datetime.strptime(file_url.split('/')[-1].split('.')[0], '%Y%m%d%H%M%S')
        if (file_dte > min_date and file_dte <= max_date):
            to_download.append(file_url)
print("{} file(s) to download since {}".format(len(to_download), min_date))
```

```python
from zipfile import ZipFile
import gzip
import io
import os
 
def download_content(url, save_path):
    with urllib.request.urlopen(url) as dl_file:
        input_zip = ZipFile(io.BytesIO(dl_file.read()), "r")
        name = input_zip.namelist()[0]
        with gzip.open(save_path, 'wb') as f:
              f.write(input_zip.read(name))

def download_to_dbfs(url):
    file_name = '{}.gz'.format(url.split('/')[-1][:-4])
    tmp_file = '{}/{}'.format("./news", file_name)
    download_content(url, tmp_file)
    # dbutils.fs.mv('file:{}'.format(tmp_file), 'dbfs:{}/{}'.format("/content", file_name))
    
n = len(to_download)
for i, url in enumerate(to_download):
    download_to_dbfs(url)
    print("{}/{} [{}]".format(i + 1, n, url))
```

```python
def unnest(df, col, reset_index=False):
    import pandas as pd
    col_flat = pd.DataFrame([[i, x] 
                       for i, y in df[col].apply(list).iteritems() 
                           for x in y], columns=['I', col])
    col_flat = col_flat.set_index('I')
    df = df.drop(col, 1)
    df = df.merge(col_flat, left_index=True, right_index=True)
    if reset_index:
        df = df.reset_index(drop=True)
    return df

df["organisation_clean2"] = df.loc[:,"organisation_clean"]
df_temp = unnest(df[["organisation_clean", "organisation_clean2"]],"organisation_clean")
# df_temp = df_temp.reset_index()
df_temp
# df_graph = unnest(df_temp[["organisation_clean", "organisation_clean2"]],"organisation_clean2")
# df_graph = df_graph[df_graph['organisation_clean'] != df_graph['organisation_clean2']]
# df_graph = df_graph.groupby(df_graph.columns.tolist(),as_index=False).size()
# df_graph = df_graph.rename(columns={"organisation_clean":"source","organisation_clean2":"target", "size":"weight"})
# import networkx as nx
# G = nx.from_pandas_edgelist(df_graph, edge_attr="weight")
```
